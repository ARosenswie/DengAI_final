# Importing packages

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns
```

# Loading the files

```python
# load the data
train_features = pd.read_csv('dengue_features_train.csv',
                             index_col=[0,2])

train_labels = pd.read_csv('dengue_labels_train.csv',
                           index_col=[0,2])
```

```python
train_features.drop(['year','week_start_date','reanalysis_dew_point_temp_k','reanalysis_min_air_temp_k','reanalysis_tdtr_k','precipitation_amt_mm','reanalysis_sat_precip_amt_mm','reanalysis_precip_amt_kg_per_m2','reanalysis_avg_temp_k','reanalysis_air_temp_k','reanalysis_max_air_temp_k'], axis=1, inplace=True)
```

```python
train_features
```

```python
# Seperate data for San Juan
sj_train_features = train_features.loc['sj']
sj_train_labels = train_labels.loc['sj']

# Separate data for Iquitos
iq_train_features = train_features.loc['iq']
iq_train_labels = train_labels.loc['iq']
```

```python
sj_train_features['total_cases'] = sj_train_labels.total_cases
iq_train_features['total_cases'] = iq_train_labels.total_cases
```
